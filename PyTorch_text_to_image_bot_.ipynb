```python
import torch
import transformers
import telebot

from diffusers import StableDiffusionPipeline
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
device
```

```python
model_id = 'XpucT/Deliberate'  #'runwayml/stable-diffusion-v1-5' #'CompVis/stable-diffusion-v1-4' # 'hassanblend/hassanblend1.4' #
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16, safety_checker=None)
pipe = pipe.to(device)
```

```python
bot = telebot.TeleBot('Please inter your authentication token')
```

```python
# Функция, обрабатывающая команду /start
@bot.message_handler(commands=["start"])
def start(m, res=False):
    bot.send_message(m.chat.id, 'Я на связи. Напиши мне на английском какую картинку ты хочешь чтобы я создал).' \
                     'Если не знаешь английский не беда deepl.com тебе в помощь:)')
```

```python
# Получение сообщений от юзера
@bot.message_handler(content_types=["text"])
def handle_text(message):
    bot.send_message(message.chat.id, message.text)
    image = pipe(message.text).images[0]
    bot.send_photo(message.chat.id, image)
```

```python
# Запускаем бота
bot.polling(none_stop=True, interval=0)
```
